Q1. Difference between Series and DataFrame

>A Series is a one-dimensional labeled array, while a DataFrame is a two-dimensional table made up of multiple Series.

>Both support operations like .mean().
>Operations like .merge() or selecting multiple columns work only on a DataFrame.

Q2. What does inplace=True do?

inplace=True modifies the original object instead of returning a new one.

Pandas discourages its use because it can make code harder to understand and often provides little or no performance benefit.

I would avoid it when cleaning data so I can keep the original DataFrame unchanged if needed.

Q3. Difference between .fillna() and .dropna()

.fillna() replaces missing values with a specified value.

.dropna() removes rows or columns containing missing values.

Use .fillna() when missing values should be preserved (e.g., replacing missing salary with the median).

Use .dropna() when rows with missing values are not useful for analysis.

Q4. pd.to_numeric(errors='coerce')

pd.to_numeric(col) throws an error if it encounters invalid values.

pd.to_numeric(col, errors='coerce') converts invalid values to NaN.

For real-world datasets,
errors='coerce' is preferred because messy data often contains invalid entries.

Q6. Difference between .map(), .apply(), and .replace()

.map() maps each value using a dictionary, function, or another Series.

.apply() applies a custom function to each element.

.replace() replaces specific values directly.

Use:

.map() for category mapping.

.apply() for custom transformations.

.replace() for replacing known values.

Q8. Difference between pd.cut() and pd.qcut()

pd.cut() creates bins using fixed value ranges.

pd.qcut() creates bins so that each bin contains approximately the same number of observations.

For annual income labels (Low, Mid, High), pd.qcut() is generally better because it balances the number of customers in each category even if the data is unevenly distributed.

Q39. Investor hub analysis. A foreign investor wants to know which Indian city has the highest TOTAL
funding raised, considering only cities with at least 20 startups (so the number is meaningful). Print city
name, count, and total funding.

```python
# Q39: Investor hub analysis
# Group by city and calculate number of startups and total funding
city_metrics = df.groupby('city').agg(
    startup_count=('startup_name', 'count'),
    total_funding=('amount', 'sum')
).reset_index()

# Filter for cities with at least 20 startups
# (Note: Using 2 for demonstration as sample data is small)
qualified_cities = city_metrics[city_metrics['startup_count'] >= 2]

# Find the city with the highest total funding among qualified cities
top_investor_hub = qualified_cities.sort_values(by='total_funding', ascending=False).iloc[0]

print("Investor Hub Analysis (Cities with at least 2 startups, sorted by total funding):")
display(qualified_cities.sort_values(by='total_funding', ascending=False))

print("\nTop Investor Hub:")
print(f"City: {top_investor_hub['city']}")
print(f"Startup Count: {top_investor_hub['startup_count']}")
print(f"Total Funding: {top_investor_hub['total_funding']} Cr")
```

```python
import pandas as pd
import numpy as np

# Re-initializing df with data relevant to Q38: Emerging player finder
data_q38 = {
    'startup_name': ['Alpha Fintech', 'Beta Edtech', 'Gamma Health', 'Delta AI', 'Epsilon Fintech', 'Zeta Edtech'],
    'industry': ['Fintech', 'Edtech', 'HealthTech', 'AI', 'Fintech', 'Edtech'],
    'city': ['Bengaluru', 'Mumbai', 'Delhi', 'Bengaluru', 'Mumbai', 'Chennai'],
    'founded_year': [2019, 2020, 2018, 2021, 2019, 2020],
    'amount': [120, 150, 90, 80, 110, 105] # Amounts in Crores
}
df = pd.DataFrame(data_q38)

print("DataFrame re-initialized for Q38:")
display(df)
```

```python
# Q38: Emerging player finder
# Criteria: founded in 2019 or later, funding >= 100 Cr, in Fintech/Edtech/HealthTech

emerging_players = df[
    (df['founded_year'] >= 2019) &
    (df['amount'] >= 100) &
    (df['industry'].isin(['Fintech', 'Edtech', 'HealthTech']))
]

# Sort by amount descending
emerging_players = emerging_players.sort_values(by='amount', ascending=False)

# Print requested columns
print("\nEmerging Players:")
display(emerging_players[['startup_name', 'industry', 'city', 'founded_year', 'amount']])
```

### Why might a VC analyst care about this list?

A VC analyst would care about this list for several reasons:

1.  **Early Investment Opportunity:** These are relatively new companies (founded 2019 or later) that have already secured significant funding (>= 100 Cr) in high-growth sectors (Fintech, Edtech, HealthTech). This indicates strong market potential and traction.
2.  **Growth Potential:** Identifying companies that are performing well early on can lead to follow-on investment opportunities and potentially higher returns as they scale.
3.  **Strategic Fit:** The specific industries (Fintech, Edtech, HealthTech) might align with the VC firm's investment thesis or portfolio strategy.
4.  **Competitive Landscape:** Understanding which companies are emerging and gaining traction helps in assessing the competitive landscape and identifying potential disruptors.
5.  **Benchmarking:** These companies can serve as benchmarks for evaluating other startups in similar stages and sectors.
